```python
#define finding the first negative numner function
def find_first_negative(lst):
    """
    Task 1
    - Create a function that finds the first negative number in a list (lst).
    - Return the first negative number if found, otherwise return "No negatives".
    - Use a while loop to implement this.
    """
    i = 0                   #initialising variable i as zero
    while i < len(lst):     # while i is less that the length of the list
        if (lst[i] < 0):    # check if the current element is negative
            return lst[i]   # return the first negative number found
        i = i + 1           # move to the next index
    return "No negatives"   #once looping all through the list and no negative no, return "no negative"


# Task 2
# Invoke the function "find_first_negative" using the following scenario:
# - [3, 5, -1, 7, -2, 8]
# - [2, 10, 7, 0]

#case1
print(find_first_negative([3,5,-1,7,-2,8]))

#case2
print(find_first_negative([2, 10, 7, 0]))
```
